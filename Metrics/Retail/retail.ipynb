+ MAU - daily active users – number of unique active users per month; it's suitable if users use your product every mponth

# Elasticity

**Price Elasticity of Demand** - the estimation of the buyer's reaction to price changes.

**Goods with Price Elastic Demand:**
- Luxury items: jewelry, delicacies.
- Goods that significantly impact the family budget: furniture, household appliances.
- Easily substitutable goods: meat, fruits.

**Goods with Price Inelastic Demand:**
- Essential items: medicines, shoes, electricity, and other energy sources.
- Goods with an insignificant impact on the family budget: pencils, toothbrushes.
- Indispensable goods: bread, light bulbs, gasoline.  

I.e. the price elasticity of demand refers to the dependency of changes in sales volume on the price. Elastic goods are those whose sales are highly sensitive to even the slightest price changes. Inelastic goods, on the other hand, are those for which demand is practically unaffected by price.
